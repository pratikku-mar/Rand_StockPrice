```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.random()

num_days = 400
volatility = 0.03
trend = 0.0006

daily_returns = np.random.normal(trend, volatility, num_days)
prices = 120 * np.cumprod(1 + daily_returns)
df = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=num_days, freq='D'),
    'Open': prices * (1 + np.random.uniform(-0.01, 0.02, num_days)),
    'High': prices * (1 + np.random.uniform(0.00, 0.02, num_days)),
    'Low': prices * (1 - np.random.uniform(0.00, 0.02, num_days)),
    'Close': prices,
    'Volume': np.random.randint(5000, 15000, num_days)
})

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], label="Stock Price")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Synthetic Stock Price Over Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df.to_csv('Synth_Stock.csv', index=False)
```
